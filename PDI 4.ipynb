```
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```
img = np.zeros((256,256,3), dtype=np.uint8)

img[0:32, :, :] = [255, 0, 0]
img[32:64, :, :] = [255, 255, 0]
img[64:96, :, :] = [0, 255, 0]

img[96:128, :, :] = [0, 255, 255]
img[128:160, :, :] = [0, 0, 255]
img[160:196, :, :] = [255, 0, 255]

img[196:228, :, :] = [255, 255, 255]
img[228:256, :, :] = [0, 0, 0]

plt.axis("off")
plt.imshow(img)
plt.show()
```

```
img = np.zeros((256,256,3), dtype=np.uint8)

for i in range(256):
  img[i, :128, :] = [255, i, i]
  img[i, 128:, :] = [255, 255-i, i]

plt.axis("off")
plt.imshow(img)
plt.show()
```

```
img = np.zeros((360,256,3), dtype=np.uint8)

for interval in range(180):
  a1 = interval*2
  a2 = (interval+1)*2
  img[a1:a2, :, :] = [interval, 255, 255]

outImg = cv2.cvtColor(img, cv2.COLOR_HSV2RGB)

plt.imshow(outImg)
plt.show()
```

```
img = cv2.imread("pathology_cll20x01.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

lower_blue = np.array([110,250,0])
upper_blue = np.array([35,100,50])

mask = cv2.inRange(img_hsv, lower_blue, upper_blue)

plt.figure(figsize=(5,5))
plt.axis("off")
plt.imshow(mask, cmap="gray")
plt.show()
```

```
img = cv2.imread("pathology_cll20x01.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

lower_blue = np.array([110,250,0])
upper_blue = np.array([130,255,255])

mask = cv2.inRange(img_hsv, lower_blue, upper_blue)

mask = cv2.GaussianBlur(mask, (3,3), 0)
result = cv2.bitwise_and(img, img, mask=mask)

plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.axis("off")
plt.imshow(img)

plt.subplot(1,2,2)
plt.axis("off")
plt.imshow(result, cmap="gray")
plt.show()
```
