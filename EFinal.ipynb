```
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Pregunta 1

```
def orange(img, img_hsv):
  
  lower = np.array([0,100,150])
  upper = np.array([180//8,255,255])

  mask = cv2.inRange(img_hsv, lower, upper)

  mask = cv2.GaussianBlur(mask, (3,3), 0)
  result = cv2.bitwise_and(img, img, mask=mask)

  return result

def white(img, img_hsv):
  
  lower = np.array([0,0,200])
  upper = np.array([180,50,255])

  mask = cv2.inRange(img_hsv, lower, upper)

  mask = cv2.GaussianBlur(mask, (3,3), 0)
  result = cv2.bitwise_and(img, img, mask=mask)

  return result 
```

# Parte naranja del pez payaso


```
for i in range(6):

  path = "nemo" + str(i) + ".jpg"
  
  img = cv2.imread(path)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  img_proc = img.copy()
  n,m,_ = img.shape
  
  img_hsv = cv2.cvtColor(img_proc, cv2.COLOR_RGB2HSV)
  
  result = orange(img, img_hsv)

  plt.figure(figsize=(15,5))
  
  plt.subplot(1,2,1)
  plt.axis("off")
  plt.imshow((img))

  plt.subplot(1,2,2)
  plt.axis("off")
  plt.imshow(result, cmap="gray")
  plt.show()
```

# Parte naranja + blanca

```
for i in range(6):

  path = "nemo" + str(i) + ".jpg"
  
  img = cv2.imread(path)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  img_proc = img.copy()
  n,m,_ = img.shape
  
  img_hsv = cv2.cvtColor(img_proc, cv2.COLOR_RGB2HSV)
  
  result = orange(img, img_hsv) + white(img, img_hsv)

  plt.figure(figsize=(15,5))
  
  plt.subplot(1,2,1)
  plt.axis("off")
  plt.imshow((img))

  plt.subplot(1,2,2)
  plt.axis("off")
  plt.imshow(result, cmap="gray")
  plt.show()
```
